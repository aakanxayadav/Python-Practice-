```python
#delete a column
import pandas as pd
info={'one':pd.Series([1,2],index=['a','b']),'two':pd.Series([1,2,3],index=['a','b','c'])}
df=pd.DataFrame(info)
print("The DataFrame:")
print(df)
print("Delete the first column:")
del df['one']
print(df)
print("Delete the another column:")
df.pop('two')
print(df)
```

```python
#select any row by passing the row label to a loc func
info={'one':pd.Series([1,2,3,4,5,6],index=['a','b','c','d','e','f']),'two':pd.Series([1,2,3,4,5,6,7,8],index=['a','b','c','d','e','f','g','h'])}
df=pd.DataFrame(info)
print(df.loc['b'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.0,2,0.01)
y1=np.sin(2*np.pi*x)
y2=1.2*np.sin(4*np.pi*x)
fig,ax=plt.subplots(1,sharex=True)
ax.plot(x,y1,x,y2,color='black')
ax.fill_between(x,y1,y2,where=y2>=y1,facecolor='lightblue',interpolate=True)
ax.fill_between(x,y1,y2,where=y2<=y1,facecolor='pink',interpolate=True)
ax.set_title('fill between where')
```

```python
plt.style.use('fivethirtyeight')
mu=50
sigma=7
x=np.random.normal(mu,sigma,size=200)
fig,ax=plt.subplots()
ax.hist(x,20)
ax.set_title('Histogram')
ax.set_xlabel('bin range')
ax.set_ylabel('frequency')
fig.tight_layout()
plt.show()
```
